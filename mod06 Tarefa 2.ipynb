```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head()
```

```python
#boxplot usando pandas
sinasc[['IDADEMAE']].boxplot()
```

```python
#boxplot usando seaborn
sns.boxplot(y='IDADEMAE', data=sinasc)
```

####  2) Faça um boxplot do peso do bebê

```python
#boxplot usando pandas
sinasc[['PESO']].boxplot()
```

```python
#boxplot usando seaborn
sns.boxplot(y='PESO', data=sinasc)
```

####  3) Faça um histograma da idade da mãe

```python
#histograma usando pandas
sinasc['IDADEMAE'].plot.hist()
```

```python
#histograma usando seaborn
sns.displot(data = sinasc['IDADEMAE']
           ,alpha = .25 
           ,element = 'step'
           ,kde = True)
```
